# Univariate Analysis

Univariate analysis examines one variable at a time to understand its distribution and characteristics.

```python
plt.figure(figsize=(12,6))

top10 = df['Make Name'].value_counts().head(10)

sns.barplot(x=top10.index, y=top10.values, palette="viridis")

plt.title("Top 10 Car Manufacturers")
plt.xlabel("Manufacturer")
plt.ylabel("Number of Models")
plt.xticks(rotation=45)

plt.show()
```

### Insight

**Why did I choose this chart?**

This chart identifies the manufacturers that have the highest number of car models in the dataset.

**Insight**

Some manufacturers offer significantly more car models than others, showing a wider product portfolio.

**Business Impact**

Manufacturers with a larger product portfolio can target more customer segments and increase their market share.

# Bivariate Analysis

Bivariate analysis examines the relationship between two variables. It helps identify patterns, trends, and correlations that can support business decisions.

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x='Engine Horsepower Hp',
    y='Trim Msrp',
    alpha=0.6
)

plt.title("Horsepower vs MSRP")
plt.xlabel("Horsepower")
plt.ylabel("MSRP")
plt.show()
```

### Insight

**Why did I choose this chart?**

To understand whether more powerful cars tend to have higher prices.

**Insight**

There is a positive relationship between horsepower and MSRP. Higher horsepower vehicles generally cost more.

**Business Impact**

Manufacturers can justify premium pricing for high-performance vehicles.

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x='Engine Size',
    y='Engine Horsepower Hp',
    alpha=0.6
)

plt.title("Engine Size vs Horsepower")
plt.show()
```

### Insight

**Why did I choose this chart?**

To explore how engine size affects horsepower.

**Insight**

**Business Impact**

This helps manufacturers balance engine size, performance, and fuel efficiency.

# Multivariate Analysis

Multivariate analysis studies the relationship among three or more variables simultaneously to uncover deeper patterns in the dataset.

```python
plt.figure(figsize=(14,10))

numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm',
    annot=False
)

plt.title("Correlation Heatmap")
plt.show()
```

### Insight

**Why did I choose this chart?**

To identify the strength and direction of relationships among all numerical variables in the dataset.

**Insight**

Strong positive correlations exist between variables such as engine size and horsepower, while fuel efficiency tends to have a negative correlation with engine performance.

**Business Impact**

Understanding these relationships helps manufacturers optimize vehicle design by balancing performance, fuel efficiency, and pricing.

```python
sns.pairplot(df[['Trim Msrp',
                 'Engine Size',
                 'Engine Horsepower Hp',
                 'Mileage Combined Mpg']])

plt.show()
```

### Insight

**Why did I choose this chart?**

To visualize pairwise relationships among important numerical features such as MSRP, engine size, horsepower, and fuel efficiency.

**Insight**

The pairplot highlights positive relationships between engine size, horsepower, and vehicle price, while showing that higher horsepower often corresponds to lower fuel efficiency.

**Business Impact**

This visualization enables manufacturers to understand multiple relationships simultaneously and make data-driven product development decisions.

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x='Engine Horsepower Hp',
    y='Mileage Combined Mpg',
    hue='Engine Fuel Type'
)

plt.title("Horsepower vs Mileage by Fuel Type")
plt.show()
```
